# Pre-processing Data

## Pembersihan Tanda Baca

```python
import re
import string
def remove_punctuation(text):
  text = re.sub(r"[^\x00-\x7f]", r"", text)
  text = re.sub(r"(\\u[0-9A-Fa-f]+)", r"", text)
  text = re.sub(r"[^A-Za-z0-9^,!.\/'+-=]", " ", text)
  return text.translate(str.maketrans("","",string.punctuation))
df["cleaning"] = df.content.apply(remove_punctuation)

def remove_number(text):
  text = re.sub(r"\b\d+\b", " ", text)
  return text.translate(str.maketrans("","",string.punctuation))
df["cleaning"] = df.cleaning.apply(remove_number)

def remove_single_char(text):
  text = re.sub(r"\b[a-zA-Z]\b", "", text)
  return text.translate(str.maketrans("","",string.punctuation))
df["cleaning"] = df.cleaning.apply(remove_single_char)

def remove_multiple_whitespace(text):
  text = re.sub('\s+',' ',text)
  return text.translate(str.maketrans("","",string.punctuation))
df["cleaning"] = df.cleaning.apply(remove_multiple_whitespace)
```

## Case Folding

```python
def case_folding(text):
  if isinstance    (text, str):
    lower_text = text.lower()
    return lower_text
  else:
    return text

df['casefolding'] = df['cleaning'].apply(case_folding)
```

## Normalisasi kata

```python
def replace_taboo_words(text, kamus_tidak_baku):
  if isinstance(text, str):
    words = text.split()
    replaced_words = []
    kalimat_baku = []
    kata_ganti = []
    kata_tidak_baku_hash = []

    for word in words:
      if word in kamus_tidak_baku:
        baku_word = kamus_tidak_baku[word]
        if isinstance(baku_word, str) and all(char.isalpha() or char.isspace() for char in baku_word):
          replaced_words.append(baku_word)
          kalimat_baku.append(baku_word)
          kata_ganti.append(word)
          kata_tidak_baku_hash.append(hash(word))
      else:
          replaced_words.append(word)
    replaced_text = ' '.join(replaced_words)
  else:
      replaced_text = ' '
      kalimat_baku = ' '
      kata_ganti = ' '
      kata_tidak_baku_hash = ' '

  return replaced_text, kalimat_baku, kata_ganti, kata_tidak_baku_hash
```

```python
kamus_data = pd.read_excel('/content/kamuskatabaku.xlsx')
kamus_tidak_baku = dict(zip(kamus_data['tidak_baku'], kamus_data['kata_baku']))
```

```python
data = pd.DataFrame(df[['source', 'content', 'cleaning', 'casefolding']])
```

```python
data['hasil normalisasi'], data['Kata_Baku'], data['Kata_Tidak_Baku'], data['Kata_Tidak_Baku_Hash'] = zip(
    *data['casefolding'].apply(lambda x: replace_taboo_words(x, kamus_tidak_baku)))
df = pd.DataFrame(data[['source', 'content', 'cleaning', 'casefolding', 'hasil normalisasi']])
```

## Tokenizing

```python
def tokenize(text):
  tokens = text.split()
  return tokens
df['tokenize'] = df['hasil normalisasi'].apply(tokenize)
```

## Stopword

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop_words = stopwords.words('indonesian')
new_stopwords = ["masingmasing", "benarbenar","dgn","gak","ga","ya","nya","yg",
                 "tolong","gabisa","mohon","aja","ngga","banget","udah","nggak",
                 "gimana","ini","gk","sih","dan","saya","karena","di","ke","untuk"]
special_words = ["terima","kasih","terimakasih","apk"]
stop_words.extend(new_stopwords)
stop_words.extend(special_words)

def remove_stopwords(text):
  return [word for word in text if word not in stop_words]

df['stopword removal'] = df['tokenize'].apply(remove_stopwords)
```

## Stemming

```python
!pip install Sastrawi
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
def stemmed_wrapper(term):
  return stemmer.stem(term)

term_dict = {}
for word in df['stopword removal']:
  for term in word:
    if term not in term_dict:
      term_dict[term] = ' '

for term in term_dict:
  term_dict[term] = stemmed_wrapper(term)

def stemming(word):
  return [term_dict[term] for term in word]
df['stemming'] = df['stopword removal'].apply(lambda x: ' '.join(stemming(x)))
```

## Sentiment

```python
from transformers import pipeline
```

```python
sentiment_classifier = pipeline(
    "text-classification",
    model="w11wo/indonesian-roberta-base-sentiment-classifier"
)

texts = [t for t in df["hasil normalisasi"].tolist() if isinstance(t, str) and t.strip() != ""]
results = sentiment_classifier(texts)
df.loc[df["hasil normalisasi"].isin(texts), "sentiment"] = [r["label"] for r in results]
df.loc[df["hasil normalisasi"].isin(texts), "score"] = [r["score"] for r in results]
```

```python
#@title Sentiment
import seaborn as sns
import matplotlib.pyplot as plt
sentiment_count = data['sentiment'].value_counts()
sns.set_style('whitegrid')

fig, ax = plt.subplots(figsize=(8,6))
ax = sns.barplot(x=sentiment_count.index, y=sentiment_count.values, palette='viridis')

plt.title('Sentimen CoreTax', fontsize=14)
total = len(data['sentiment'])

for i, count in enumerate(sentiment_count.values):
  percentage = f'{100 * count / total:.2f}%'
  ax.text(i, count + 0.10, f'{count}\n({percentage})', ha='center', va='bottom', fontsize=8)

plt.show()
```

```python
#@title Sentiment Setiap Sumber
plt.figure(figsize=(10,6))
ax = sns.countplot(data=data, x="source", hue="sentiment", palette="viridis")
group_totals = data.groupby("source")["sentiment"].count()

for p in ax.patches:
    height = p.get_height()
    if height == 0:
        continue

    x = p.get_x() + p.get_width() / 2
    source = ax.get_xticklabels()[int(x)].get_text()
    total = group_totals[source]
    percentage = 100 * height / total
    ax.annotate(f'{percentage:.1f}%',
                (x, height),
                ha='center', va='bottom',
                fontsize=9, color='black')

plt.title("Distribusi Sentiment Setiap Sumber")
plt.xlabel("Source")
plt.ylabel(" ")
plt.tight_layout()
plt.show()
```

```python
#@title WordCloud & Frequency
from wordcloud import WordCloud
positive_texts = ' '.join(data[data['sentiment'] == 'positive']['stemming'])
negative_texts = ' '.join(data[data['sentiment'] == 'negative']['stemming'])
neutral_texts = ' '.join(data[data['sentiment'] == 'neutral']['stemming'])

print(f"  - Positive texts: {len(positive_texts.split())} words")
print(f"  - Negative texts: {len(negative_texts.split())} words")
print(f"  - Neutral texts: {len(neutral_texts.split())} words")


fig, axes = plt.subplots(1, 3, figsize=(24, 7))


if len(positive_texts.strip()) > 0:
    wordcloud_pos = WordCloud(
        width=600,
        height=400,
        background_color='white',
        colormap='Greens',
        max_words=100
    ).generate(positive_texts)

    axes[0].imshow(wordcloud_pos, interpolation='bilinear')
    axes[0].set_title('WordCloud - Sentimen POSITIF', fontsize=16, color='green')
    axes[0].axis('off')
else:
    axes[0].text(0.5, 0.5, 'Tidak ada data positif', ha='center', va='center', fontsize=14)
    axes[0].axis('off')

if len(negative_texts.strip()) > 0:
    wordcloud_neg = WordCloud(
        width=600,
        height=400,
        background_color='white',
        colormap='Reds',
        max_words=100
    ).generate(negative_texts)

    axes[1].imshow(wordcloud_neg, interpolation='bilinear')
    axes[1].set_title('WordCloud - Sentimen NEGATIF', fontsize=16, color='red')
    axes[1].axis('off')
else:
    axes[1].text(0.5, 0.5, 'Tidak ada data negatif', ha='center', va='center', fontsize=14)
    axes[1].axis('off')


if len(neutral_texts.strip()) > 0:
    wordcloud_neu = WordCloud(
        width=600,
        height=400,
        background_color='white',
        colormap='Greys',
        max_words=100
    ).generate(neutral_texts)

    axes[2].imshow(wordcloud_neu, interpolation='bilinear')
    axes[2].set_title('WordCloud - Sentimen NETRAL', fontsize=16, color='gray')
    axes[2].axis('off')
else:
    axes[2].text(0.5, 0.5, 'Tidak ada data netral', ha='center', va='center', fontsize=14)
    axes[2].axis('off')

plt.tight_layout()
plt.show()


from collections import Counter

def get_top_words(text, n=15):
    if not text or len(text.strip()) == 0:
        return []
    words = text.split()
    words = [w for w in words if len(w) > 3]
    return Counter(words).most_common(n)

top_positive = get_top_words(positive_texts, 15)
top_negative = get_top_words(negative_texts, 15)
top_neutral = get_top_words(neutral_texts, 15)


fig, axes = plt.subplots(1, 3, figsize=(22, 6))


if top_positive:
    pos_words, pos_counts = zip(*top_positive)
    axes[0].barh(pos_words, pos_counts, color='#2ecc71', alpha=0.8)
    axes[0].set_title('Top 15 Kata - Sentimen POSITIF', fontsize=14)
    axes[0].set_xlabel('Frekuensi', fontsize=12)
    axes[0].invert_yaxis()
    axes[0].grid(axis='x', alpha=0.3)
else:
    axes[0].text(0.5, 0.5, 'Tidak ada data', ha='center', va='center', fontsize=12)
    axes[0].set_title('Top 15 Kata - Sentimen POSITIF', fontsize=14)


if top_negative:
    neg_words, neg_counts = zip(*top_negative)
    axes[1].barh(neg_words, neg_counts, color='#e74c3c', alpha=0.8)
    axes[1].set_title('Top 15 Kata - Sentimen NEGATIF', fontsize=14)
    axes[1].set_xlabel('Frekuensi', fontsize=12)
    axes[1].invert_yaxis()
    axes[1].grid(axis='x', alpha=0.3)
else:
    axes[1].text(0.5, 0.5, 'Tidak ada data', ha='center', va='center', fontsize=12)
    axes[1].set_title('Top 15 Kata - Sentimen NEGATIF', fontsize=14)


if top_neutral:
    neu_words, neu_counts = zip(*top_neutral)
    axes[2].barh(neu_words, neu_counts, color='#95a5a6', alpha=0.8)
    axes[2].set_title('Top 15 Kata - Sentimen NETRAL', fontsize=14)
    axes[2].set_xlabel('Frekuensi', fontsize=12)
    axes[2].invert_yaxis()
    axes[2].grid(axis='x', alpha=0.3)
else:
    axes[2].text(0.5, 0.5, 'Tidak ada data', ha='center', va='center', fontsize=12)
    axes[2].set_title('Top 15 Kata - Sentimen NETRAL', fontsize=14)

plt.tight_layout()
plt.show()
```

# TF-IDF Ranking

```python
import pandas as pd
data = pd.read_csv('/content/Hasil Preprocessing CoreTax(Play Store, Youtube, Twitter & TikTok).csv')
data = data.dropna()
print("Data setelah drop missing value", data.isnull().sum())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

data['stemming'] = data['stemming'].fillna('')
df_list = data['stemming'].tolist()

df_tfidf = TfidfVectorizer()
X = df_tfidf.fit_transform(df_list)

print('Bentuk Transformasi Feauture Extraxtion', X.shape)

feature_names = df_tfidf.get_feature_names_out()

tfidf_sums = X.sum(axis=0)
tfidf_df = pd.DataFrame({'term': feature_names, 'tfidf_sum': tfidf_sums.tolist()[0]})
tfidf_df = tfidf_df.sort_values(by='tfidf_sum', ascending=False)

print("\nTop 10 Terms by Total TF-IDF Score")
print(tfidf_df.head(10))

print("\nIDF Values")
idf_values = df_tfidf.idf_
idf_df = pd.DataFrame({'term': feature_names, 'IDF': idf_values})
print(idf_df.head(10))

print("\nTF-IDF Values for the first 5 documents")
for i in range(min(5, len(df_list))):
  first_document_tfidf = pd.DataFrame(X[i].T.todense(), index=feature_names, columns=["TF-IDF"])
  print(f"\nDocument {i+1} TF-IDF Values:")
  print(first_document_tfidf.sort_values(by=["TF-IDF"], ascending=False).head(10))
```

```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

neg_data = data[data['sentiment'] == 'negative']
texts = neg_data['hasil normalisasi'].fillna('').tolist()
#texts = data['hasil normalisasi'].fillna('').tolist()

embedding_model = SentenceTransformer("distiluse-base-multilingual-cased-v2")

topic_model = BERTopic(
    language="indonesian",
    embedding_model=embedding_model,
    n_gram_range=(1, 2),
    min_topic_size=20
)

topics, probs = topic_model.fit_transform(texts)
topic_model.get_topic_info()
```

```python
topic_model.save("bertopic_coretax_model")
```

```python
df_results = pd.DataFrame({
    "text": texts,
    'sentiment': neg_data['sentiment'],
    "topic": topics,
    "probability": probs,
    "topic_name": [topic_model.get_topic(t) for t in topics],
    "topic_words": [topic_model.get_topic(t)[0] for t in topics],
    "topic_score": [topic_model.get_topic(t)[1] for t in topics],
    "source": neg_data['source']
})

df_results
```
